```
# (ORIGIN, MONTH) LOOKUP TABLE
import pandas as pd

filenames = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

custom_df = pd.DataFrame()

for name in filenames:
  path = "/content/drive/My Drive/PROJECT 1001/Data3/new_data_" + name + ".csv"
  df = pd.read_csv(path)
  df['ORIGIN_COMBINED'] = list(zip(df.ORIGIN, df.MONTH))

  feature_df = df.query('ARR_DEL15==1').groupby('ORIGIN_COMBINED')['ARR_DEL15'].count()
  feature_df = pd.concat([feature_df, df.query('ARR_DEL15==0').groupby('ORIGIN_COMBINED')['ARR_DEL15'].count()], axis=1)
  feature_df.columns = ['DELAY_YES', 'DELAY_NO']
  feature_df['TOTAL'] = feature_df['DELAY_YES'] + feature_df['DELAY_NO']
  feature_df['DELAY_TO_NO_DELAY_RATIO'] = feature_df['DELAY_YES'] / feature_df['DELAY_NO']
  feature_df['DELAY_PROB'] = feature_df['DELAY_YES'] / feature_df['TOTAL']

  minimum = feature_df["DELAY_TO_NO_DELAY_RATIO"].min()
  q25 = feature_df["DELAY_TO_NO_DELAY_RATIO"].quantile(0.25)
  median = feature_df["DELAY_TO_NO_DELAY_RATIO"].quantile(0.5)
  q75 = feature_df["DELAY_TO_NO_DELAY_RATIO"].quantile(0.75)
  maximum = feature_df["DELAY_TO_NO_DELAY_RATIO"].max()

  feature_df['DelayScore'] = 0
  score_list = []
  for x in range(len(feature_df)):
    ratio = feature_df["DELAY_TO_NO_DELAY_RATIO"].iloc[x]
    if ratio <= minimum:
      score = 1
    elif ratio <= q25:
      score = 2
    elif ratio <= median:
      score = 3
    elif ratio <= q75:
      score = 4
    else:
      score = 5
    score_list.append(score)
  feature_df['DelayScore'] = score_list
  custom_df = custom_df.append(feature_df)
```

```
# WEATHER LOOKUP TABLE
# Source: https://www.ncei.noaa.gov/access/search/data-search/global-summary-of-the-day
import pandas as pd
filenames = ['ORD.csv',   'CLT.csv',   'MDW.csv',   'ATL.csv',   'BOS.csv',   'DFW.csv',   'DEN.csv',   'IAH.csv',   'LAX.csv',   'LAS.csv',
             'DTW.csv',   'HOU.csv',   'DCA.csv',   'IAD.csv',   'PHX.csv',   'SAN.csv',   'SFO.csv',   'SEA.csv',   'AZA.csv',   'LGA.csv',
             'PHL.csv',   'EWR.csv',   'MSP.csv',   'JFK.csv',   'MCO.csv']
custom_df = pd.DataFrame()

for name in filenames:
  path = "/content/drive/My Drive/PROJECT 1001/Data5/" + name
  df = pd.read_csv(path)
  df['STATION'] = name.replace('.csv', '')
  df['AIRPORT_DATE'] = list(zip(df.STATION, df.DATE))
  custom_df = custom_df.append(df)
```
