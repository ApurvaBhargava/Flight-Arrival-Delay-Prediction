```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

###Delay Trends for a given Origin Airport 

```
klist = [(i, j) for i in range(3) for j in range(4)]
```

```
filenames = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
fig, axes = plt.subplots(figsize=(24,10), nrows=3, ncols=4, sharey=True)

for index in range(12):
  path = "/content/drive/My Drive/PROJECT 1001/Data2/new_" + filenames[index] + ".csv"
  i, j = klist[index]
  df = pd.read_csv(path)
  df2 = pd.pivot_table(df, index='ORIGIN', values=['TAXI_OUT'], aggfunc=np.mean)
  if index==0:
    temp = df2
  df2.sort_values(by='ORIGIN').plot(kind='bar', ax=axes[i, j])
  temp.sort_values(by='ORIGIN').plot(kind='line', ax=axes[i, j], color='r', rot=90)
```

```
airport_list = ['SEA', 'CLT', 'PHX', 'HOU', 'DEN', 'EWR', 'PHL', 'SAN', 'DCA', 'BOS', 'SFO', 'LGA', 'ORD', 'MDW', 'LAS', 'DFW', 'LAX', 'ATL', 'MCO', 'JFK',
                'MSP', 'IAD', 'IAH', 'DTW', 'AZA']
filenames = ['feb', 'jan', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
fig, axes = plt.subplots(figsize=(24,10), nrows=3, ncols=4, sharey=True)

for index in range(12):
  path = "/content/drive/My Drive/PROJECT 1001/Data2/new_" + filenames[index] + ".csv"
  i, j = klist[index]
  df = pd.read_csv(path)
  df2 = pd.pivot_table(df, index='ORIGIN', values=['CARRIER_DELAY'], aggfunc=np.mean)
  if index==0:
    temp = df2
  df2.reindex(airport_list).plot(kind='bar', ax=axes[i, j])
  temp.reindex(airport_list).plot(kind='line', ax=axes[i, j], color='r', rot=90)
```
