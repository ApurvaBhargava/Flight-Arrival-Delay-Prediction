###Grouping scheduled departure time in groups and plotting delay trends

```
filenames = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
timelist = ['00:00:00', '01:00:00', '02:00:00', '03:00:00', '04:00:00', '05:00:00', '06:00:00', '07:00:00', '08:00:00', '09:00:00', '10:00:00', '11:00:00',
            '12:00:00', '13:00:00', '14:00:00', '15:00:00', '16:00:00', '17:00:00', '18:00:00', '19:00:00', '20:00:00', '21:00:00', '22:00:00', '23:00:00',
            '24:00:00']
timelist = ['00:00:00', '03:00:00', '06:00:00', '09:00:00', '12:00:00', '15:00:00', '18:00:00', '21:00:00', '24:00:00']
klist = [(i, j) for i in range(3) for j in range(4)]
```

```
fig, axes = plt.subplots(figsize=(24,18), nrows=3, ncols=4, sharey=True)
index = 0

for file in filenames:
  m, n = klist[index]
  index += 1
  path = "/content/drive/My Drive/PROJECT 1001/Data2/new_" + file + ".csv"
  df = pd.read_csv(path)
  df = df[np.isfinite(df['DEP_DELAY'])]
  df = df[np.isfinite(df['ARR_DELAY'])]
  df['SUB_DEP_DELAY'] = pd.to_datetime(df.DEP_DELAY, unit='m').dt.strftime('%H:%M')
  df['DEP_TIME'] = df['DEP_TIME'].astype('int')
  df['DEP_TIME'] = df['DEP_TIME'].astype('str')
  df['DEP_TIME'] = df['DEP_TIME'].replace('2400', '0000')
  df['DEP_TIME'] = df['DEP_TIME'].str.pad(width=4, side='left', fillchar='0')
  df['DEP_TIME'] = pd.to_datetime(df['DEP_TIME'], format='%H%M', errors='coerce').dt.strftime('%H:%M')
  df = df[df['DEP_TIME']!='NaT']
  # Conversion to datetime
  df['SUB_DEP_DELAY'] = df['SUB_DEP_DELAY'].apply(lambda x: datetime.datetime.strptime(x, '%H:%M'))
  df['DEP_TIME'] = df['DEP_TIME'].apply(lambda x: datetime.datetime.strptime(x, '%H:%M'))
  # Calculation of scheduled departure time
  df['SCHED_DEP_TIME'] = df['DEP_TIME'] - df['SUB_DEP_DELAY']
  mean_arr_del15 = np.zeros((len(timelist)-1,))
  if file=='jan':
    ref_line = mean_arr_del15
  for i in range(len(timelist)-1):
    df2 = df[(df['SCHED_DEP_TIME']>=timelist[i]) & (df['SCHED_DEP_TIME']<=timelist[i+1])]
    mean_arr_del15[i] = df2['ARR_DEL15'].mean()
  
  #fig, ax1 = plt.subplots(figsize=(5, 5))
  axes[m,n].bar(timelist[0:8], mean_arr_del15)
  axes[m,n].plot(timelist[0:8], ref_line, color='r')
  axes[m,n].set_xticklabels(timelist[0:8], rotation=90)
  axes[m,n].set_xlabel('Time group lower bound')
  axes[m,n].set_ylabel('Mean count of arrival delays')
```
