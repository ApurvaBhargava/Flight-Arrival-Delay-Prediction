```
import pandas as pd
import numpy as np
import time
import datetime
```

####Dropping cancellation and diversion related columns as well as rows that contain null values for target/important attributes

```
filenames = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

for name in filenames:
  path = "/content/drive/My Drive/PROJECT 1001/Data2/new_" + name + ".csv"
  print(path)
  df = pd.read_csv(path)
  #print("Original Shape:", df.shape)
  #print("No of rows to be removed:", df.loc[df['LONGEST_ADD_GTIME'].notna()].shape)
  df = df[df['LONGEST_ADD_GTIME'].isna()]
  df = df[df['CANCELLATION_CODE'].isna()]
  df = df[np.isfinite(df['ARR_DEL15'])]
  df = df[np.isfinite(df['DEP_DEL15'])]
  #print("New Shape:", df.shape)
  #print("No. of non-NaN rows:", df.loc[df['LONGEST_ADD_GTIME'].notna()].shape)
  new_name = 'new_data_'+name+'.csv'
  df.to_csv(new_name)
  !cp '$new_name' 'drive/My Drive/PROJECT 1001/Data3/'
  print()
```

####Adding SCHED_DEP_TIME (scheduled departure time column)

```
filenames = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

for name in filenames:
  path = "/content/drive/My Drive/PROJECT 1001/Data3/new_data_" + name + ".csv"
  print(path)
  df = pd.read_csv(path)
  df = df[np.isfinite(df['DEP_DELAY'])]
  df = df[np.isfinite(df['ARR_DELAY'])]
  df['SUB_DEP_DELAY'] = pd.to_datetime(df.DEP_DELAY, unit='m').dt.strftime('%H:%M')
  df['DEP_TIME'] = df['DEP_TIME'].astype('int')
  df['DEP_TIME'] = df['DEP_TIME'].astype('str')
  df['DEP_TIME'] = df['DEP_TIME'].replace('2400', '0000')
  df['DEP_TIME'] = df['DEP_TIME'].str.pad(width=4, side='left', fillchar='0')
  df['DEP_TIME'] = pd.to_datetime(df['DEP_TIME'], format='%H%M', errors='coerce').dt.strftime('%H:%M')
  # Conversion to datetime
  df['SUB_DEP_DELAY'] = df['SUB_DEP_DELAY'].apply(lambda x: datetime.datetime.strptime(x, '%H:%M'))
  df['DEP_TIME'] = df['DEP_TIME'].apply(lambda x: datetime.datetime.strptime(x, '%H:%M'))
  # Calculation of scheduled departure time by subtracting delay from actual departure time
  df['SCHED_DEP_TIME'] = df['DEP_TIME'] - df['SUB_DEP_DELAY']
  df['SCHED_DEP_TIME'] = df['SCHED_DEP_TIME'] - pd.to_timedelta(df['SCHED_DEP_TIME'].dt.days, unit='d')
  new_name = 'fin_'+name+'.csv'
  df.to_csv(new_name)
  !cp '$new_name' 'drive/My Drive/PROJECT 1001/Data4/'
  print()
```
