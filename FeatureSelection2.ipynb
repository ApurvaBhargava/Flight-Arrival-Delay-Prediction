```
import pandas as pd
import numpy as np

path = "/content/drive/My Drive/PROJECT 1001/processed_dataset.csv"
data = pd.read_csv(path)
data.drop(columns='Unnamed: 0', inplace=True)
```

```
X = data.drop(['ARR_DEL15', 'ARR_DELAY'], axis=1)
y = data['ARR_DEL15']
```

```
X.columns
```

##**Feature Importances using DecisionTreeClassifier**

```
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
%matplotlib inline

clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)

# Code here
feature_names = ['QUARTER', 'MONTH', 'DAY_OF_MONTH', 'DAY_OF_WEEK', 'AIR_TIME',
       'DISTANCE', 'PRCP', 'WDSP', 'VISIB', 'SLP', 'DELAY_TO_NO_DELAY_RATIO',
       'DELAY_PROB', 'DelayScore', 'DELAY_NO_DELAY_RATIO_DEST',
       'DELAY_PROB_DEST', 'DelayScore_DEST', 'TIME_GROUP_NUM',
       'TIME_ARR_DEL15', 'TIME_ARR_DELAY', 'TIME_DEP_DEL15', 'TIME_DEP_DELAY',
       'CARRIER_TAXI_OUT', 'CARRIER_CARRIER_DELAY', 'CARRIER_NAS_DELAY',
       'ORIGIN_TAXI_OUT', 'ORIGIN_CARRIER_DELAY', 'ORIGIN_NAS_DELAY',
       'ORIGIN_LATE_AIRCRAFT_DELAY']
feature_importances = clf.feature_importances_

# Sort feature importances and accordingly, the feature_names
feature_importances, feature_names = (list(t) for t in zip(*sorted(zip(feature_importances, feature_names))))

# Plot bar chart
plt.figure(figsize=(12, 8))
plt.bar(np.arange(len(feature_importances)), feature_importances, align='center', alpha=0.5)
plt.xticks(np.arange(len(feature_importances)), feature_names, rotation='vertical')
plt.xlabel('Feature')
plt.ylabel('Importance Value')

plt.show()
```
